## 2. 경로 / 기타 인자 설정

### 데이터 경로 설정

````python
# Project directory
PROJECT_DIR = '/home/sugnjin/Desktop/apple_classification'
os.chdir(PROJECT_DIR)

# Data directory
DATA_DIR = os.path.join(PROJECT_DIR, 'DATA')
TRAIN_IMG_DIR = os.path.join(DATA_DIR, 'train')
TRAIN_LABEL_DIR =  os.path.join(DATA_DIR, 'train.csv')
TEST_IMG_DIR =  os.path.join(DATA_DIR, 'test')
TEST_LABEL_DIR =  os.path.join(DATA_DIR, 'test.csv')
SAMPLE_DIR = os.path.join(DATA_DIR, 'sample_submission.csv')
````

````python
# File count
print(len(os.listdir(TRAIN_IMG_DIR)))
print(len(os.listdir(TEST_IMG_DIR)))
````

### 결과 저장 경로 설정

````python
# Time 
kst = timezone(timedelta(hours=9)) # kst = utc + 9
train_serial = datetime.now(kst).strftime('%Y%m%d_%H%M%S')

# Record Directory
RECORD_DIR = os.path.join(PROJECT_DIR, 'RECORD', 'RESULT', train_serial)
# os.mkdir(RECORD_DIR)
os.makedirs(RECORD_DIR, exist_ok = True)
````

### 시드 설정

````python
# Seed
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
````

### 디바이스 설정

````python
# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
````

## 3. 데이터 불러오기 및 분할

### Train 라벨

````python
traindf = pd.read_csv(TRAIN_LABEL_DIR)
traindf.head()
````

### Train 이미지 예시

````python
example_index = random.randrange(len(traindf)) # 0,1,2...add
example_imageid = traindf['img_id'][example_index]
example_image_path = os.path.join(TRAIN_IMG_DIR, example_imageid)
example_image = Image.open(example_image_path)

plt.imshow(example_image)

````

### Train / Validation 분할

````python
# 80% train, 20% validation
train, val = train_test_split(traindf, test_size=0.2, random_state=RANDOM_SEED, shffle=True)

# train / validation split dataframe save
train.to_csv(os.path.join(DATA_DIR, 'train_label.csv'), index=False)
val.to_csv(os.path.join(DATA_DIR, 'val_label.csv'), index=False)
````
